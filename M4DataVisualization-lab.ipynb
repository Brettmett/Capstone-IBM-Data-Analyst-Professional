```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM Master
"""
df_stacked = pd.read_sql_query(QUERY,conn)
df_stacked
# filtering for the Agegroup:
df_30_35 = df_stacked[(df_stacked["Age"]>29)&(df_stacked["Age"]<36)]
df_30_35
```

```python
# grouping by Age:
sns.set_theme(style="white")
df_30_35.groupby(["Age"]).median()[["WorkWeekHrs","CodeRevHrs"]].plot(kind="bar", stacked=True)
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT ConvertedComp, Age
FROM Master
WHERE (Age >29 And Age<36)
"""
df_conv = pd.read_sql_query(QUERY,conn)
df_conv
```

```python
# grouping by Age:
df_grouped = df_conv.groupby(["Age"]).median()["ConvertedComp"].reset_index()
df_grouped.columns= ["Age", "ConvertedComp_Median"]
df_grouped
```

```python
# creating LineChart:
sns.set_theme(style="whitegrid")
sns.lineplot(df_grouped, x= "Age", y= "ConvertedComp_Median")
plt.ylabel("Yearly Income Median")
plt.title("Income by Age (30-35)")
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
# your code goes here
QUERY = """
SELECT MainBranch
FROM Master
"""
df_mainbr = pd.read_sql_query(QUERY,conn)
df_mainbr = df_mainbr["MainBranch"].value_counts().reset_index()
df_mainbr.columns=["MainBranch", "Count"]
```

```python
# Plotting horizantal Barchart:
sns.set_theme(style="whitegrid")
sns.set_color_codes("pastel")
sns.barplot(df_mainbr, x="Count", y = "MainBranch")
plt.ylabel("Main Profession")
```

## Extract information from the given web site
You will extract the data from the below web site: <br> 


```python
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data you need to scrape is the **name of the programming language** and **average annual salary**.<br> It is a good idea to open the url in your web broswer and study the contents of the web page before you start to scrape.


Import the required libraries


```python
# Your code here
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

Download the webpage at the url


```python
#your code goes here
data = requests.get(url).text
data
```

Create a soup object


```python
#your code goes here
soup = BeautifulSoup(data,"lxml")
print(soup.prettify())                  # Displays in Html-Format for better Readebility..               
```

Scrape the `Language name` and `annual average salary`.


```python
#Searching for Table
table = soup.find("table")
list_languages= []
list_salaries = []
# Iterating over table rows:
for i in table.find_all("tr"):
    language_col = (i.find_all("td")[1].text)
    avg_annual_salary_col = (i.find_all("td")[3].text)

    # appending to lists:
    list_languages.append(language_col)
    list_salaries.append(avg_annual_salary_col)
    df_data = pd.DataFrame(list_languages, list_salaries).reset_index()

# First row as column Name:
df_data.columns = ["Average Annual Salary", "Language"]
# dropping first Row, since i
df_data.drop(index=0,inplace=True)
df_data
```

Save the scrapped data into a file named *popular-languages.csv*


```python
# your code goes here
df_data.to_csv("/Users/christianbretthauer/Desktop/git_public/Capstone-IBM-Data-Analyst-Professional/data/popular_languages.csv")
```
